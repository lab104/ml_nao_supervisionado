# 1. Introdução

Este notebook trata da criação de um modelo de detecção de fraudes em uma base pública de cartões de crédito que serve como um exemplo do tratamento de *conjuntos de dados desbalanceados*. A base em questão contém cerca de 280 mil transações, das quais apenas 492 são fraude. 

> The dataset contains transactions made by credit cards in September 2013 by European cardholders.
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions (https://www.kaggle.com/mlg-ulb/creditcardfraud)

Este é um problema interessante por dois motivos:
* Indicadores de Acurácia e F1 apresentam resultados bons, mesmo quando o modelo é ruim. 
* Modelos de aprendizado supervisionado não convergem bem se o conjunto de dados é desbalanceado. 

Para tratar este problema foram adotadas as seguintes estratégias:
* Foram adotados indicadores de Sensitividade (percentual da fraude total identificada pelo modelo) e Precisão (percentual das fraudes previstas que eram efetivamente fraude)
* No conjunto de treinamento, as transações fraudulentas foram repetidas mais de 500 vezes para que o volume total de fraudes e não fraudes fosse mais ou menos o mesmo. Tal balanceamento foi efetuado apenas na base de treinamento. A base de testes ficou inalterada para peritir a avaliação do modelo sem distorções. 
* Foi utilizada treinada uma rede neural usando um grid search para encontrar os melhores hiperparâmetros. 

Algumas variações foram testadas, mas não trouxeram resultados melhores:
* Treinar o modelo com 100% da fraude e uma amostra aleatória de mesmo tamanho da não fraude;
* Mudança da função de ativação para tanh e logistic;
* Treinado o modelo com o valor como vem;
* Treinado o modelo com o valor quebrado em colunas binárias de faixa de valores.

# 3. Preparação dos dados para o modelo

O código a seguir:
* Divide o conjunto de dados em dois:
  * **train** com 80% dos registros;
  * **test** com 20% dos registros.
* Balanceia o conjunto de treinamento. A técnica usada foi duplicar o conjunto de não fraude até que fraude e não fraude correspondam a mais ou menos 50% do conjunto de dados.
* Define as "features" que serão usadas no treinamento.

```python
def block():
    
    global df
    global X_train, X_test, X_validation
    global y_train, y_test, y_validation    
    global df_features
    
    df_features = [
         'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9','V10',
        'V11','V12','V13','V14','V15','V16','V17','V18','V19','V20',
        'V21','V22','V23','V24','V25','V26','V27','V28',
        'vr_pad'
        #,  'vr_f1','vr_f2','vr_f3', 'vr_f4', 'vr_f5', 'vr_f6',
        #'vr_f7','vr_f8','vr_f9','vr_f10','vr_f11','vr_f12'       
    ]
    
    from sklearn.model_selection import train_test_split
            
    # first split (df into train and test)
    X_train, X_test = train_test_split(df, test_size=0.2, stratify=df['Class'], random_state=564654)
    
    # second split (tain into test and validation)
    # X_train, X_validation = train_test_split(X_train, test_size=0.2, stratify=X_train['Class'])
      
    # ballance classes    
    X_train_fraude = X_train[ X_train['Class'] == 1 ]
    X_train_naofraude = X_train[ X_train['Class'] == 0 ]
    
    # STRATEGY 1: Duplicate fraud until reach the size of the normal transactions
    for i in range( int( len(X_train_naofraude)/len(X_train_fraude) ) ):
        X_train = X_train.append( X_train_fraude )
    
    X_train = X_train.sample(frac=1, random_state=564654).reset_index(drop=True)    
    
    # STRATEGY 2: All fraud + Sample of the normal transactions
    #X_train = X_train_fraude.sample(n=10*len(X_train_naofraude),random_state=564654, replace=True)
    #X_train = X_train.append( X_train_naofraude.sample(n=10*len(X_train_fraude),random_state=564654) )
    
       
    # third split: (classes)
    y_train = X_train.pop( 'Class' )
    y_test = X_test.pop( 'Class' )
    #y_validation = X_validation.pop( 'Class')
        
    print(f'Training shape: {X_train.shape} {y_train.shape}' )
    #print(f'Validation shape: {X_validation.shape} {y_validation.shape}' )
    print(f'Test shape: {X_test.shape} {y_test.shape}' )
         
block()
```

# 4. Configura o modelo e treina

Neste exemplo, foi usado um [GridSearch][1] para avaliar diversas combinações de hiperparametros para a rede neural [MLPClassifier][2]. 

Toma muito tempo para avaliar os diferentes parâmetros, mas vale a pena pois ajuda a desenvolver a sensibilidade do que funciona bem para este tipo de problema. Nós rodamos o grid várias vezes para refinar os melhores resultados encontrados de cada interação e as rodadas anteriores estão comentadas no código. Abaixo do código a tabela com os último grid executado com os hiperparametros em ordem do melhor score. 

[1]: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
[2]: https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

```python
def block():
    
    global model
    
    import sklearn.neural_network 
    import sklearn.model_selection
    import pandas as pd
        
    model = sklearn.neural_network.MLPClassifier(
                solver='adam',            #default='adam'
                learning_rate_init=0.001, #default=0.001 step size when updating weights
                alpha=0.0001,             #default=0.0001 
                hidden_layer_sizes=(16, 16), 
                random_state=564654,
                activation='relu',        #default='relu'
                max_iter=100,
                early_stopping=True
    )
    
    #grid = {        
    #   'learning_rate_init': [ 0.01, 0.001, 0.0001 ],
    #   'hidden_layer_sizes': [ (8), (8,8), (8,8,8) ],
    #   'solver': ['adam'],
    #   'activation': ['logistic','relu','tanh'],
    #   learning_rate': ['constant','invscaling','adaptive']     
    #}
    
    #grid = {        
    #    'learning_rate_init': [ 0.1, 0.01, 0.02 ],
    #    'hidden_layer_sizes': [ (8,8), (16,16) ],
    #    'alpha': [ 0.0001, 0.001, 0.01 ],
    #    'solver': ['adam'],
    #    'activation': ['relu'],
    #}
    
    #grid = {        
    #    'learning_rate_init': [ 0.1, 0.01, 0.02 ],
    #    'hidden_layer_sizes': [ (16,16), (32,16), (16,32) ],
    #    'alpha': [ 0.0001, 0.001, 0.01 ],
    #    'solver': ['adam'],
    #    'activation': ['relu'],
    #}
    
    grid = {        
        'learning_rate_init': [ 0.01, 0.02 ],
        'hidden_layer_sizes': [ (32,16), (64,16), (64,32) ],
        'alpha': [ 0.0001, 0.001 ],
        'solver': ['adam'],
        'activation': ['relu'],
    }
    
    # define the score as the harmonic average between sensitivity and precision
    def score( y_true, y_pred ):
        tn, fp, fn, tp = sklearn.metrics.confusion_matrix(y_true,y_pred).ravel()
        sensitivity = tp/(fn+tp)
        precision = tp/(fp+tp)
        print( f'calculated sensitivity={sensitivity} precision={precision}' )        
        #return (2*sensitivity*precision)/(sensitivity+precision)
        return (sensitivity+precision)/2
        
    #model = sklearn.model_selection.RandomizedSearchCV( 
    model = sklearn.model_selection.GridSearchCV( 
        model, 
        grid,
        cv=2,                     # only two folds for each option
        verbose=3,       
        n_jobs=2, pre_dispatch=2, # parallel processes
        scoring=sklearn.metrics.make_scorer(score)
    )        
    
    model.fit( X_train[ df_features ], y_train )          
          
block()
```

```python
def block():    
    import pandas as pd
    
    #print( model.best_estimator_ )
    rel = pd.DataFrame( model.cv_results_ )
    rel = rel.sort_values( by='rank_test_score', ascending=True )    
    rel = rel.drop( columns=['params','std_fit_time', 'std_score_time', 'split0_test_score', 'split1_test_score', 'std_test_score', 'mean_fit_time' ] )         
    rel.insert( 0, 'rank_test_score', rel.pop( 'rank_test_score' ) )
    rel.insert( 1, 'mean_test_score', rel.pop( 'mean_test_score' ) )
    #rel = rel.loc[ ('rank_test_score', 'mean_test_score', 'mean_score_time'), : ]        
    return rel    
        
block()
```

# 5. Avaliação dos resultados

Na [matriz de confusão][2] (abaixo):
* A diagonal verde representa as previsões corretas feitas pelo modelo;
* A diagonal vermelha representa as previsões incorretas feita pelo modelo;

Neste tipo de problema o objetivo é detectar o máximo de *verdadeiros positivos* (tp) com um mínimo de *falsos negativos* (fn) e um mínimo de *falsos positivos* (fp). Para refletir isto foi selecionado o seguinte conjunto de indicadores:
* **[Sensitividade][1] (True Positive Rate or Recall)** = `tp / (tp+fn)`
* **[Precisão][1] (Positive Predictive Value)** = `tp / (tp+fp)`.

Neste tipo de problema, é comum que a *[Acurácia][3]* (`(tp+tn/tp+tn+fp+fn)` ) apresente resultados próximos a 100% mesmo quando o modelo não tenha um bom poder preditivo. Um simples *experimento mental* pode ilustrar isto: um modelo que alerte apenas classe=0 com os mesmos dada matriz de confusão abaixo teria uma acurácia de 99,82%, mesmo sem ter valor preditivo. O *F-Score* é a média harmônica da *Acurácia* e do *Recall*, portanto herda parte da distorção da *Acurácia*. É importante ressaltar que estas limitações são particulares ao tipo de problema que está sendo avaliado e portanto não devem ser tomadas como regra geral.

[1]: https://en.wikipedia.org/wiki/Precision_and_recall
[2]: https://en.wikipedia.org/wiki/Confusion_matrix
[3]: https://towardsdatascience.com/accuracy-precision-recall-or-f1-331fb37c5cb9?gi=ddfdb7bbd07f

```python
def block():
    
    import matplotlib
    import matplotlib.pyplot as plt
    from sklearn.metrics import confusion_matrix
    #from sklearn.metrics import roc_curve
    import seaborn as sns
    import pandas as pd
    
    # calculate the predictions
    y_pred = model.predict(X_test[ df_features ])
        
    class_names = ['Não Fraude\nPrevista','Fraude\nPrevista']
        
    confusion_mtx = pd.DataFrame( 
        columns=class_names,        
        data=confusion_matrix( y_test, y_pred )
    )    
    confusion_mtx['Total'] = confusion_mtx['Fraude\nPrevista'] + confusion_mtx['Não Fraude\nPrevista']
    confusion_mtx = confusion_mtx.append(confusion_mtx.sum(numeric_only=True), ignore_index=True)
    confusion_mtx = confusion_mtx.rename( index={ 0:'Não Fraude Real', 1:'Fraude Real', 2:'Total' } )
    
    fig = plt.figure(figsize=(8,8))    

    sns.set(font_scale=1.5)
    
    c = sns.heatmap([[2,1,0],[1,2,0],[0,0,0]], 
                    annot=[
                        ['\nTN\n\n(true\nnegative)','\nFP\n\n(false\npositive)',''],
                        ['\nFN\n\n(false\nnegative)','\nTP\n\n(true\npositive)','\n\n\n(total\nfraud)'],
                        ['\n\n','\n\n\n(total\npredicted)','\n\n(obs)']], 
                    cbar=False, 
                    fmt='',
                    linewidths=2,
                    linecolor='#000000',
                    cmap=matplotlib.colors.ListedColormap(['#ffffff','#FFa0a0','#a0FFa0'])
                   )
    c = sns.heatmap(confusion_mtx,                     
                    annot=True, 
                    fmt='g', 
                    cbar=False, 
                    cmap=matplotlib.colors.ListedColormap(['#ffffff00']),
                    linewidths=2,                    
                    linecolor='#000000'
                   )
    
    #c.set(xlabel='Classe Prevista',ylabel='Classe Real')
    c.set(title='Matriz de Confusão')
    c.tick_params(right=True, bottom=True, labelright=True, labelleft=False, labeltop=False,rotation=0)


    fp = confusion_mtx.iloc[0,1] 
    fn = confusion_mtx.iloc[1,0]
    tn = confusion_mtx.iloc[0,0]
    tp = confusion_mtx.iloc[1,1] 
    
    acc       = (tp+tn)/(tp+tn+fp+fn)
    recall    = tp/(tp+fn)
    f1        = (2*acc*recall)/(acc+recall)
    
    fdr  = fp/(fp+tp)
    fnr  = fn/(fn+tp)
    
    tpr = tp/(fn+tp)
    ppv  = tp/(fp+tp)
    
    print( f'      Accuracy={acc:05.2f}' )
    print( f'       F Score={f1:05.2f}' )
    print( f'   Sensitivity={tpr*100.0:05.2f}% of the real fraud was detected' )   
    print( f'     Precision={ppv*100.0:05.2f}% of the predicted fraud was real fraud' )    
        
    plt.text( s=f'Sensitividade:\n{tpr*100.0:05.2f}% of', x=2.05, y=1.25, size='small', color='#808080' )
    plt.text( s=f'Precisão:\n{ppv*100.0:05.2f}% of', ha='right', x=1.95, y=2.25, size='small', color='#808080' )
    
    plt.show()
    
block()
```

# Apêndice 1 - Revisão de conceitos

### PCA

A [transformação PCA][1] consiste em uma "redução de dimensionalidade" usualmente é utilizada para simplificar e reduzir o número de colunas de um dataset. 

Entre outras coisas, atributos fortemente correlacionados acabarão resumidos. Isto é feito plotando cada coluna original como um eixo e buscando linhas perpendiculares entre si que correspondam as direções de maiores variações dos dados. Tais linhas se tornarão os novos eixos e os dados serão projetados nestes novos eixos. 

Um efeito colateral da redução de dimensionalidade é difícil reverter tal transformação, mas ela mantém características suficientes dos dados originais para permitir muitos estudos serem efetuados, assim serve como uma forma de anonimizar a base. 

### Leiaute das transações de cartão de crédito.

[O leiaute ISO8583][2] é um padrão de mercado muito usado nas redes de cartão de débito e crédito em nível mundial e apresenta os tipos de campos que devem ser encontrados:  
* conta;
* código da loja;
* código da maquineta;
* tipo de loja;
* endereço da loja;
* tipo de senha;
* cartão utilizados;
* situação do terminal;
* etc

### StratifiedKFold Cross Validation

O [GridSearch][3] tem opção para usar o [StratifiedKFold][4]. Em essência este médoto subdivide o conjunto de treinamento em fatias, usa [uma das fatias para validação e as demais fatias para treinamento][5], conforme ilustrado abaixo para um kfold com cv=4 (4 fatias).
      
     iteracao 1: |V1|T2|T3|T4|  treino: T2,T3,T4, validação: T1
     iteracao 2: |T1|V2|T3|T4|  treino: T1,T3,T4, validação: T2
     iteracao 3: |T1|T2|V3|T4|  treino: T1,T2,T4, validação: T3
     iteracao 4: |T1|T2|T3|V4|  treino: T1,T2,T3, validação: T4 
     
Esta é uma boa forma de trabalho pois permite usar 100% da base de treino sem perder um pequeno percentual para a validação. Isto é especialmente relevante para as bases desbalanceadas em que a classe de interesse usualmente tem poucas observações.
